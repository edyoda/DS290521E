```python
class A:
    def __init__(self):
        print('in A')
    def eat(self):
        print('A is eating')
        
class B(A):
    def __init__(self):
        print('in B')
    def eat(self):
        print('B is eating')
        super().eat()
        
class C(A):
    def __init__(self):
        print('in C')
    def eat(self):
        print('C is eating') 
        
class D(B):
    def __init__(self):
        print('in D')
    def eat(self):
        print('D is eating')
        super().eat()
        
class E(C):
    def __init__(self):
        print('in E')
    def eat(self):
        print('E is eating')
        
class X(D,E):
    def __init__(self):
        print('is X')
        super().eat()
    def eat(self):
        print('X is eating')
        
obj = X()
```

```python
class paytm:
    list_of_user={}
    list_of_login_user={}
    
    def  __init__(self):
        print("Welcome to Paytm")
        

    def sign_up(self,name,email,password):
        self.name=name
        self.email=email
        self.password=password
        paytm.list_of_user[self.name]=[self.email,self.password]
        


    def login(self,email,password):
        if self.name in paytm.list_of_user:
            if(self.email==email and self.password==password):
                paytm.list_of_login_user[self.name]=[self.email,self.password]
                print("Login successful")
            else:
                print("Incorrect login email and password")
        else:
            print("You don't have any account ,please sign-up")

    def logout(self):
        if self.name in paytm.list_of_login_user:
            print("you are logged out")
            paytm.list_of_login_user.pop(self.name)
        else:
            print("you are not logged in")
```
